Outliers present in a number of the features

```python
# Removing the upper the 99 percentile to remove outliers, of all columns aside from geographical ones.
outlier_cols = [col for col in X_train.columns if col not in ['Latitude', 'Longitude']]
for col in outlier_cols:
    upper_bound = X_train[col].quantile(0.99)
    X_train = X_train[X_train[col] <= upper_bound]

X_train.hist(figsize=(12, 8), bins=30)
plt.show()
```
